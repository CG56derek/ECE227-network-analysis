```python
import networkx as nx
import time
from preprocess import G_fb, G_enron, G_erdos
from parallel_worker import parallel_betweenness

def extract_top_nodes_parallel(G, network_name):
    print(f"\n[{network_name}]")
    start_time = time.time()
    
    # 1. Degree Centrality
    deg_cent = nx.degree_centrality(G)
    top_5_deg = sorted(deg_cent.items(), key=lambda x: x[1], reverse=True)[:5]
    print(f"Degree Top 5: {[n for n, v in top_5_deg]}")
    
    # 2. Eigenvector Centrality
    try:
        eig_cent = nx.eigenvector_centrality(G, max_iter=1000)
        top_5_eig = sorted(eig_cent.items(), key=lambda x: x[1], reverse=True)[:5]
        print(f"Eigenvector Top 5: {[n for n, v in top_5_eig]}")
    except nx.PowerIterationFailedConvergence:
        print("Eigenvector Centrality: Cannot converge")

    # 3. Closeness Centrality 
    close_cent = nx.closeness_centrality(G)
    top_5_close = sorted(close_cent.items(), key=lambda x: x[1], reverse=True)[:5]
    print(f"Closeness Top 5: {[n for n, v in top_5_close]}")
        
    # 4. Betweenness Centrality
    bet_cent = parallel_betweenness(G)
    top_5_bet = sorted(bet_cent.items(), key=lambda x: x[1], reverse=True)[:5]
    print(f"Betweenness Top 5: {[n for n, v in top_5_bet]}")


if __name__ == '__main__':
    extract_top_nodes_parallel(G_fb, "Facebook")
    extract_top_nodes_parallel(G_erdos, "Erdös (Collaboration)")
    extract_top_nodes_parallel(G_enron, "Enron Email")
```

```python
import networkx as nx
from preprocess import G_fb, G_erdos, G_enron

def test_node_hypothesis(G, network_name, target_nodes):
    print(f"\n[{network_name}]")
    print(f"{'Node ID':<10} | {'Degree':<8} | {'Clustering Coeff (Cv)':<25} | {'Avg Neighbor Degree (Knn)':<25}")
    print("-" * 75)
    
    for node in target_nodes:
        if node not in G:
            continue
            
        degree = G.degree(node)
        
        clustering = nx.clustering(G, node)
        
        neighbors = list(G.neighbors(node))
        if len(neighbors) > 0:
            avg_neighbor_degree = sum(G.degree(n) for n in neighbors) / len(neighbors)
        else:
            avg_neighbor_degree = 0.0
            
        print(f"{node:<10} | {degree:<8} | {clustering:<25.4f} | {avg_neighbor_degree:<25.2f}")

if __name__ == '__main__':
    fb_nodes = [107, 1912] 
    
    erdos_nodes = [430, 421, 125]
    
    enron_nodes = [5038, 136]

    test_node_hypothesis(G_fb, "Facebook", fb_nodes)
    test_node_hypothesis(G_erdos, "Erdös (Collaboration)", erdos_nodes)
    test_node_hypothesis(G_enron, "Enron Email", enron_nodes)
```
